### 2. Monthly distribution analysis of number of rainy days

#### England

```python

# Convert 'date' column to datetime format
data_en['date'] = pd.to_datetime(data_en['date'])

# Create subplots
fig, axs = plt.subplots(2, 6, figsize=(16, 6))

# Define rainfall bins
rainfall_bins = [5, 10, 15, 20, 25, 30, 35]

# Iterate over each month
for month in range(1, 13):
    # Filter data for the current month
    month_data = data_en[data_en['date'].dt.month == month]
    
    # Group data into bins
    counts, _ = np.histogram(month_data['Rainfall'], bins=rainfall_bins)
    
    # Calculate the x-coordinates for the bars in the subplot
    x = np.arange(len(rainfall_bins) - 1)
    
    # Calculate the position of the subplot in the 2x6 grid
    row = (month - 1) // 6
    col = (month - 1) % 6
    
    # Plot the bar chart in the corresponding subplot
    # Plot the bar chart in the corresponding subplot
    bars = axs[row, col].bar(x, counts, width=0.6,color='g', align='center', tick_label=rainfall_bins[:-1])
    axs[row, col].set_title(pd.to_datetime('2023-' + str(month) + '-1').strftime('%B'))
    axs[row, col].set_ylabel('Number of Days')
    axs[row, col].set_xlabel('Rainfall Range (mm)')
   # Update y-axis limits and ticks
    axs[row, col].set_ylim(0, 300)
    axs[row, col].set_yticks(np.arange(0, 301, 50))
    
    # Print the number of days above each bar
    for bar, count in zip(bars, counts):
        axs[row, col].annotate(str(count), xy=(bar.get_x() + bar.get_width() / 2, bar.get_height()), xytext=(0, 3),
                               textcoords='offset points', ha='center', va='bottom', fontsize=9)

plt.suptitle("Analysis of rain days and range by Month (1981-2023) : England", fontsize=16)
plt.tight_layout()
plt.show()
```

### 4. Daily rainfall analysis for all locations for the rainy months(Oct,Nov, Dec and Jan)

```python
import pandas as pd
import matplotlib.pyplot as plt

# Convert the 'Date' column to datetime format
data_en['date'] = pd.to_datetime(data_en['date'])

# Filter data for January, October, November, and December
filtered_data_en = data_en[(data_en['date'].dt.month >= 10) | (data_en['date'].dt.month == 1)]
filtered_data_wl = data_wl[(data_wl['date'].dt.month >= 10) | (data_wl['date'].dt.month == 1)]
filtered_data_sc = data_sc[(data_sc['date'].dt.month >= 10) | (data_sc['date'].dt.month == 1)]
filtered_data_ir = data_ir[(data_ir['date'].dt.month >= 10) | (data_ir['date'].dt.month == 1)]

# Calculate the yearly averages for each day
yearly_avg_en = filtered_data_en.groupby(filtered_data_en['date'].dt.strftime('%m-%d')).mean()
yearly_avg_wl = filtered_data_wl.groupby(filtered_data_wl['date'].dt.strftime('%m-%d')).mean()
yearly_avg_sc = filtered_data_sc.groupby(filtered_data_sc['date'].dt.strftime('%m-%d')).mean()
yearly_avg_ir = filtered_data_ir.groupby(filtered_data_ir['date'].dt.strftime('%m-%d')).mean()

# Reorder the index so that January appears at the end
january_avg_en = yearly_avg_en.loc['01-01':'01-31']
yearly_avg_en = yearly_avg_en.drop(january_avg_en.index)
yearly_avg_en = pd.concat([yearly_avg_en, january_avg_en])

january_avg_wl = yearly_avg_wl.loc['01-01':'01-31']
yearly_avg_wl = yearly_avg_wl.drop(january_avg_wl.index)
yearly_avg_wl = pd.concat([yearly_avg_wl, january_avg_wl])


january_avg_sc = yearly_avg_sc.loc['01-01':'01-31']
yearly_avg_sc = yearly_avg_sc.drop(january_avg_sc.index)
yearly_avg_sc = pd.concat([yearly_avg_sc, january_avg_sc])


january_avg_ir = yearly_avg_ir.loc['01-01':'01-31']
yearly_avg_ir = yearly_avg_ir.drop(january_avg_ir.index)
yearly_avg_ir = pd.concat([yearly_avg_ir, january_avg_ir])


# Plot the graph
plt.figure(figsize=(14, 4))
plt.plot(yearly_avg_en.index, yearly_avg_en['Rainfall'], color='g',label="England")
plt.plot(yearly_avg_wl.index, yearly_avg_wl['Rainfall'], color='b',label="Wales")
plt.plot(yearly_avg_sc.index, yearly_avg_sc['Rainfall'], color='m',label="Scotland")
plt.plot(yearly_avg_ir.index, yearly_avg_ir['Rainfall'], color='y',label='Northern Ireland')

plt.legend(frameon=False)
plt.xlabel('Date (MM-DD)')
plt.ylabel('Yearly Average Rainfall')
plt.title('Yearly Average Rainfall for October, November, December and January')
plt.xticks(rotation=75)
plt.tight_layout()

# Show the plot
plt.show()
```

### 6. Seasonal rainfall analysis for all locations

```python
import pandas as pd
import matplotlib.pyplot as plt
import calendar


# Convert the 'Date' column to datetime format
data_en['date'] = pd.to_datetime(data_en['date'])
data_wl['date'] = pd.to_datetime(data_wl['date'])
data_sc['date'] = pd.to_datetime(data_sc['date'])
data_ir['date'] = pd.to_datetime(data_ir['date'])

# Define the seasons and their corresponding months
seasons = {
    'Spring': [3, 4, 5],
    'Summer': [6, 7, 8],
    'Autumn': [9, 10, 11],
    'Winter': [12, 1, 2]
}

# Function to calculate seasonal mean rainfall
def calculate_seasonal_mean(data):
    seasonal_means = {}
    for season, months in seasons.items():
        filtered_data = data[data['date'].dt.month.isin(months)]
        seasonal_mean = np.round_(filtered_data['Rainfall'].sum()/42.5,1)
        seasonal_means[season] = seasonal_mean
    return seasonal_means

# Calculate seasonal mean rainfall for each location
seasonal_means_en = calculate_seasonal_mean(data_en)
seasonal_means_wl = calculate_seasonal_mean(data_wl)
seasonal_means_sc = calculate_seasonal_mean(data_sc)
seasonal_means_ir = calculate_seasonal_mean(data_ir)

locations = ['England', 'Wales', 'Scotland', 'Northern Ireland']
seasons = list(seasons.keys())
data = [
    [seasonal_means_en[season] for season in seasons],
    [seasonal_means_wl[season] for season in seasons],
    [seasonal_means_sc[season] for season in seasons],
    [seasonal_means_ir[season] for season in seasons]
]

# add patterns to bar plots
hatch=['///','***','|||','....']

# Plot the bar chart
fig, ax = plt.subplots(figsize=(12, 5))
width = 0.2
for i, season in enumerate(seasons):
    bars= ax.bar([x + width*i for x in range(len(locations))], [d[i] for d in data], width=width, color='#78a8f5', edgecolor='black', label=season,  hatch=hatch[i])
    for bar in bars:
        height = bar.get_height()
        ax.annotate(f'{height:.2f}', xy=(bar.get_x() + bar.get_width() / 2, height), xytext=(0, 3), textcoords='offset points', ha='center', va='bottom',fontsize=10)

ax.set_xticks([x + width*1.5 for x in range(len(locations))])
ax.set_xticklabels(locations, fontsize=14)
ax.set_ylabel('Mean Rainfall (mm)', fontsize=14)
#ax.set_title('Statistics of seasonal rainfall in four regions from 1981-2023', fontsize=16)
ax.legend(frameon=False)

# Adjust top margin
plt.subplots_adjust(top=1)

# Remove right and top spines
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
plt.savefig("seasonal rainfall 1981-2023.png", dpi=1200, quality=90)

plt.show()
```
